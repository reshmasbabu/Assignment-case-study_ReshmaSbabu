```python
data.head(10)
```

```python
#Counting the classification of species
plt.title('Count of Species')
sns.countplot(data["Classification"])
plt.show()
```

DATA INTERPRETATION 1 
* Each species ( Iris-setosa, Iris-versicolor, Iris virginic) has 50 counts each.
* Meaning all the species are well balanced in the given dataset

```python
#Comparison between various species based on SL and SW
plt.figure(figsize=(17,9))
plt.title('Comparison between various species based on SL and SW')
sns.scatterplot(data['SL'],data['SW'],hue =data['Classification'],s=50)
plt.show()
```

DATA INTERPRETATION 2
* Iris Setosa species has smaller SL but higher SW.
* Virginica has larger SL and smaller SW
* Versicolor holds middle value.

```python
#Analysis using pair plots
sns.pairplot(data,hue="Classification",height=4)
```

DATA INTERPRETATION 3

* Setosa has both low PL and PW
* Versicolor has both average PL and PW
* Virginica has both high PL and PW
